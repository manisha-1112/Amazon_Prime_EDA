## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
import pandas as pd
import ast

# Load datasets
titles = pd.read_csv("titles.csv")
credits = pd.read_csv("credits.csv")

# ----------------------------------------------------
# 1. Convert 'genres' from stringified list to actual list
titles['genres'] = titles['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# 2. Merge the datasets on 'id'
df = pd.merge(titles, credits, on='id', how='left')

# ----------------------------------------------------
# 3. Clean the merged data

# Fill missing 'description' with placeholder text
df['description'] = df['description'].fillna('No description available')

# Fill missing 'age_certification' with 'Unknown'
df['age_certification'] = df['age_certification'].fillna('Unknown')

# Fill missing 'seasons' with 0 and convert to int
df['seasons'] = df['seasons'].fillna(0).astype(int)

# Convert release year to int safely
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').fillna(0).astype(int)

# Convert scores and popularity to float
df['imdb_score'] = pd.to_numeric(df['imdb_score'], errors='coerce')
df['tmdb_score'] = pd.to_numeric(df['tmdb_score'], errors='coerce')
df['tmdb_popularity'] = pd.to_numeric(df['tmdb_popularity'], errors='coerce')

# Clean 'credits' columns
df.dropna(subset=['name', 'role'], inplace=True)
df['role'] = df['role'].str.upper()

# Optional: Fill missing names if needed
df['name'] = df['name'].fillna('Unknown')

# ----------------------------------------------------
# Final Clean Dataset Ready
print("Cleaned and merged dataset is ready.")
df.head()
```

### What all manipulations have you done and insights you found?



### **Data Manipulations Performed:**

* **Parsed and extracted** year_added and month_added from the date_added field for time-based trend analysis.
* **Cleaned and split** the duration column into two new features: duration_int and duration_type for better numerical analysis.
* **Renamed** listed_in to genre for clarity and standardization.
* **Handled missing values** in key fields like country, director, and cast by replacing them with "Unknown".
* **Converted stringified fields** like genres and production_countries into proper list formats.
* **Standardized numeric fields** (seasons, release_year, imdb_score, tmdb_score, etc.) for consistency.
* **Cleaned the credits data** by removing nulls and duplicates and standardized the role field to uppercase.



### **Insights Found:**

* Many entries were missing director and cast, especially in TV Show-type content.
* Most content was released or added **after 2015**, with a **peak in 2019–2020**.
* duration_type shows **movies use "min"**, while **TV shows use "Season"** predominantly.
* Multiple entries have very low imdb_score or tmdb_score, suggesting potential quality concerns or lesser-known titles.
